## Loading the data

```python
from pandas import Series, DataFrame #Series and DataFrame are used alot so imported direclty to local namespace
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import time
import pylab

filename = 'trip_data_1.csv'
start = time.time()
df0 = pd.read_csv(filename, parse_dates = True) 

#For interest - measuring the time taken to load in the CSV file
duration = time.time() - start
print duration
```

## Limiting Data to JFK Pickups

```python
df = df0[df0.pickup_latitude < 40.65453]
df = df[df.pickup_latitude > 40.640]
df = df[df.pickup_longitude < -73.775]
df = df[df.pickup_longitude > -73.795]
df.pickup_datetime = df.pickup_datetime.apply(lambda d: datetime.strptime(d,'%Y-%m-%d %H:%M:%S'))
```

## Cleaning the data

```python
#Eliminating datapoints where the latitude or longitude are wrong:
df.dropna() #drops rows with missing data

df = df[(df.dropoff_longitude < -10) & (df.dropoff_longitude > -150)]
df = df[(df.dropoff_latitude > 10) & (df.dropoff_latitude < 150)]
df = df[(df.pickup_longitude < -10)]
df = df[(df.pickup_latitude > 10) & (df.pickup_latitude < 150)]
```

## Visualize Pick-up location data

```python
#  SCATTER PLOT - Pickup
fig = plt.figure(1,(14, 12),facecolor="white")
ax1 = fig.add_subplot( 1, 1, 1)
ax1.scatter(df['pickup_longitude'], df['pickup_latitude'], color="blue",alpha = 0.006)
ax1.ticklabel_format(axis='both', style='plain')

#Format the x and y axis units
def form3(x, pos):
    """ This function returns a string with 3 decimal places, given the input x"""
    return '%.3f' % x
from matplotlib.ticker import FuncFormatter
formatter = FuncFormatter(form3)
ax1.xaxis.set_major_formatter(FuncFormatter(formatter))
ax1.yaxis.set_major_formatter(FuncFormatter(formatter))

plt.xlabel('Pickup Longitude')
plt.ylabel('Pickup Latitude')
plt.xlim([-73.795, -73.775])
plt.ylim([40.640, 40.655])
plt.title('Taxi Pickup Locations at JFK Airport')
```

```python
#Define variable giving the day of the month the pickups occured on 
df['pickup_day'] = df.pickup_datetime.apply(lambda d: d.strftime("%x"))
```

```python
#Count the number of pickups that happen each day
pickup_count = list(df['pickup_day'].groupby(df.pickup_day).apply(lambda d: len(d)))
```

```python
#Create a dataframe for the daily pickup and weather data
Day_by_Day = pd.DataFrame()
Day_by_Day['pickup_day'] = np.sort(list(set(df['pickup_day'])))
Day_by_Day['Day_of_Week'] = Day_by_Day.pickup_day.apply(lambda d: datetime.strptime(d,'%m/%d/%y').strftime("%w")).astype('int')
Day_by_Day['No_pickups'] = pickup_count
#Day_by_Day = Day_by_Day.drop(Day_by_Day.index[0])
```

```python
#Load in the weather data downloaded as a CSV
filename = "421423.csv"
df2 = pd.read_csv(filename, parse_dates = True) 
```

#### SNOW - Snowfall (mm)
#### TMIN - Minimum temperature (tenths of degrees C)
#### TMAX - Maximum temperature (tenths of degrees C)
#### SNWD - Snow depth (mm)
#### PRCP - Precipitation (tenths of mm)

```python
#Attach weather data to the daily pickup dataframe
i = 32
Day_by_Day['Rain_Prcp'] = df2.PRCP[:i]
Day_by_Day['SNWD'] = df2.SNWD[:i]
Day_by_Day['SNOW'] = df2.SNOW[:i]
Day_by_Day['TMAX'] = df2.TMAX[:i]
#Day_by_Day['TMIN'] = df2.TMIN[:i]
Day_by_Day.Rain_Prcp += 1
log_rain = [x for x in np.log(Day_by_Day.Rain_Prcp.values)]
Day_by_Day.Rain_Prcp += -1
```

```python
#### LINEAR REGRESSION
from sklearn import linear_model
regr = linear_model.LinearRegression()
#Assessing Rain Precipitation as a Predictor of No. of Taxi Pickups
for i in (1,3,4,5,6):
    X = Day_by_Day.ix[:,[i]]
    y  = Day_by_Day.ix[:,2]
    regr.fit(X, y)
    R_Squared = regr.score(X, y)
    print 'R-Squared for Linear Regresion with %r is %.3f'% (Day_by_Day.columns[i], R_Squared)
```

```python
#### LINEAR REGRESSION
X = Day_by_Day.ix[:,[1,3,4,5,6]]
y  = Day_by_Day.ix[:,2]
regr.fit(X, y)
R_Squared = regr.score(X, y)
print 'R-Squared for Linear Regresion is %.3f'% R_Squared
```

```python
Lin_Models = [linear_model.LinearRegression, linear_model.Ridge, linear_model.Lasso, linear_model.ElasticNet]
Lin_Models += [linear_model.MultiTaskLasso, linear_model.LassoLars, linear_model.OrthogonalMatchingPursuit]
Lin_Models += [linear_model.BayesianRidge, linear_model.ARDRegression]
Lin_Models += [linear_model.SGDClassifier, linear_model.Perceptron, linear_model.PassiveAggressiveClassifier]

for regression in Lin_Models:
    mod = regression().fit(X, y)
    R_Squared = mod.score(X, y)
    print 'R-Squared for %r \t\t\t= %.5f'% (regression, R_Squared)
    
```

#### Conclusion: Daily weather and Day-of-week information is not a good predictor of total number of daily taxi pickups at JFK. 
#### None of the various sklearn linear model classifiers did any better than the standard linear regression model
