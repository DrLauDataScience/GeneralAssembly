```python
#Function to load the individual csv files storing the traffic data and total the number of taxi pickups for each hour of each day
def hourly(filename, xlim = [-73.795, -73.775], ylim = [40.637, 40.653]):
    df = pd.read_csv(filename, parse_dates = True) 
    df.rename(columns=lambda x: x.replace(' ', ''), inplace=True)
    #data columns are offset and need adjusting
    df['dropff_longitude']  = df['rate_code']
    df['dropff_latitude']  = df['store_and_fwd_flag']
    del df['rate_code']
    del df['store_and_fwd_flag']
    df.pickup_datetime = df.pickup_datetime.apply(lambda d: datetime.strptime(d,'%Y-%m-%d %H:%M:%S').strftime('%x %H:00'))
    df_day = pd.DataFrame()
    x = df['pickup_datetime'].groupby(df['pickup_datetime']).apply(lambda d: len(d))
    df_day['datehour'] = x.index
    df_day['no_pickups'] = x.values
    #Look at subset of total pickups at JFK Airport
    df_jfk = pd.DataFrame()
    df_jfk = df[(df.pickup_longitude > xlim[0]) & (df.pickup_longitude < xlim[1]) & (df.pickup_latitude > ylim[0]) & (df.pickup_latitude < ylim[1])]
    x2 = df_jfk['pickup_datetime'].groupby(df_jfk['pickup_datetime']).apply(lambda d: len(d))
    df_day_jfk = pd.DataFrame()
    df_day_jfk['datehour'] = x2.index
    df_day_jfk['jfk_pickups'] = x2.values
    #Merge total and jfk pickup counts
    df_day = pd.merge(df_day, df_day_jfk, how = 'outer')
    df_day.fillna(0, inplace = True)
    return df_day
```

```python
#Use glob to read in data from CSV files and save totals to a new csv file
import glob
df_taxi = pd.DataFrame()
start = time.time()
for file in glob.glob(dir + "*.csv"):
    print file
    df_taxi = df_taxi.append(hourly(file))
duration = time.time() - start
print duration
df_taxi.to_csv(dir + 'taxi_hourly.csv')
```
