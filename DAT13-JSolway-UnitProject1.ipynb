## Cleaning the data

```python
#Eliminating datapoints where the latitude or longitude are wrong:
df.dropna() #drops rows with missing data

df = df[(df.dropoff_longitude < -10)]
df = df[(df.dropoff_latitude > 10) & (df.dropoff_latitude < 150)]
df = df[(df.pickup_longitude < -10)]
df = df[(df.pickup_latitude > 10) & (df.pickup_latitude < 150)]
```

## Comparison of pickup and dropoff locations

```python
#  SCATTER PLOT - Pickup
fig = plt.figure(1,(17, 13),facecolor="white")
ax1 = fig.add_subplot( 2, 2, 1)
ax1.scatter(df['pickup_longitude'], df['pickup_latitude'], color="blue",alpha = 0.2)
plt.xlabel('Pickup Longitude')
plt.ylabel('Pickup Latitude')
plt.xlim([-74.7, -73.2])
plt.ylim([40.4, 41.2])
plt.title('Pickup Locations')
#  SCATTER PLOT - Dropoff
ax2 = fig.add_subplot( 2, 2, 2)
ax2.scatter(df['dropoff_longitude'], df['dropoff_latitude'], color = "red",alpha = 0.5)
plt.xlabel('Dropoff Longitude')
plt.ylabel('Dropoff Latitude')
plt.xlim([-74.7, -73.2])
plt.ylim([40.4, 41.2])
plt.title('Dropoff Locations')
```

# Analysis of one medallion cab route

```python
df1 = df[(df.medallion == df.medallion[7])]
i = 0
longitude = []
latitude = []
for i in range(len(df1)):
	longitude.append(df1.iloc[i,10])
	longitude.append(df1.iloc[i,12])
	latitude.append(df1.iloc[i,11])
	latitude.append(df1.iloc[i,13])
fig2 = plt.figure(1,facecolor="white")
ax3 = fig2.add_subplot( 1, 1, 1)
ax3.plot(longitude, latitude, color="green",linestyle ='--', marker ='o')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('One Medallion Route')
```

## Average Trip Time

```python
# HIST PLOT
fig = plt.figure(1,(14,5),facecolor="white")
ax = fig.add_subplot( 1, 1, 1)
ax.hist(df['trip_time_in_secs']/60, bins = range(0,60,2), color="brown")
plt.title('Trip Time')
plt.xlabel('Trip time / mins')
plt.xlim([0,60])
```

## Variation in pick-up location with time-of-day

```python
#Adding columns to dataframe for hour of pick-up and day of pick-up
df['pickup_hour'] = df.pickup_datetime.apply(lambda d: datetime.strptime(d,'%Y-%m-%d %H:%M:%S').hour)
```

```python
fig = plt.figure(1,(20, 40),facecolor="white")
for i in range(24):
    ax = fig.add_subplot( 12, 2, i)
    ax.scatter(df.pickup_longitude[df.pickup_hour == i], df.pickup_latitude[df.pickup_hour == i], color="grey",alpha = 0.2)
    time_label = (str(i)+':00')
    plt.xlim([-74.7, -73.2])
    plt.ylim([40.4, 41.2])
    plt.title(time_label)
```

```python
df['pickup_day'] = df.pickup_datetime.apply(lambda d: datetime.strptime(d,'%Y-%m-%d %H:%M:%S').strftime("%A"))
```

```python
fig = plt.figure(1,(20, 40),facecolor="white")
dayofweek = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
for i in range(7):
    ax = fig.add_subplot( 4, 2, i)
    ax.scatter(df.pickup_longitude[df.pickup_day == dayofweek[i]], df.pickup_latitude[df.pickup_day == dayofweek[i]], color="purple",alpha = 0.2)
    plt.xlim([-74.7, -73.2])
    plt.ylim([40.4, 41.2])
    plt.title(dayofweek[i])
```
